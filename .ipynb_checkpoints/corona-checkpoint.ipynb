```python

data6=ascii.read('double.csv')
x1=data6['col1']
x2=data6['col2']
data3=ascii.read('after_100.csv')
data3
plt.rcParams["figure.figsize"] = (20,20)
data3
font = {'family' : 'serif',
        'weight' : 'normal',
        'size'   : 28}
import matplotlib.dates as mdates
india=data3['col4']
italy=data3['col8']
iran=data3['col40']
usa=data3['col20']
japan=data3['col12']
uk=data3['col16']
france=data3['col44']
aus=data3['col48']
pak=data3['col36']
spain=data3['col24']
germany=data3['col52']
sk=data3['col28']
sin=data3['col32']
x=data3['col1']
plt.gca().tick_params(direction='in', length=20, width=2)
minorLocator2 = AutoMinorLocator(5)
plt.gca().xaxis.set_minor_locator(minorLocator2)
plt.gca().tick_params(direction='in', which='minor', length=8, width=2)
plt.gca().set_yscale('log')

plt.plot(x,india,color='r',linewidth=9,label='India',marker='o',markersize=16)
plt.plot(x,italy,linewidth=4,label='Italy',marker='o',markersize=10)
plt.plot(x,usa,color='b',linewidth=4,label='USA',marker='o',markersize=10)
plt.plot(x,iran,color='g',linewidth=4,label='Iran',marker='o',markersize=10)
plt.plot(x,france,color='c',linewidth=4,label='France',marker='o',markersize=10)
plt.plot(x,sin,color='m',linewidth=4,label='Singapore',marker='o',markersize=10)
plt.plot(x,sk,color='y',linewidth=4,label='South Korea',marker='o',markersize=10)
plt.plot(x,japan,color='k',linewidth=4,label='Japan',marker='o',markersize=10)
plt.plot(x,uk,color='c',linewidth=4,linestyle='--',label='UK',marker='o',markersize=10)
plt.plot(x,aus,color='m',linewidth=4,linestyle='--',label='Australia',marker='o',markersize=10)
plt.plot(x,pak,color='y',linewidth=4,linestyle='--',label='Pakistan',marker='o',markersize=10)
plt.plot(x,germany,color='k',linewidth=4,linestyle='--',label='Germany',marker='o',markersize=10)
plt.plot(x,spain,color='b',linewidth=4,linestyle='--',label='Spain',marker='o',markersize=10)


plt.legend(prop={'size': 24})
plt.gcf().autofmt_xdate()
plt.gca().set_xlabel(r'Days')
plt.gca().set_ylabel('Confirmed cases after 100 MARK ')
plt.gca().set_xlim(-1,30)
plt.gca().set_ylim(80,100000)
plt.plot(x1*2,x2,color='b',linewidth=4,alpha=0.2)
plt.plot(x1*3,x2,color='b',linewidth=4,alpha=0.2)
plt.plot(x1*4,x2,color='b',linewidth=4,alpha=0.2)
plt.plot(x1*5,x2,color='b',linewidth=4,alpha=0.2)
plt.plot(x1*6,x2,color='b',linewidth=4,alpha=0.2)
plt.plot(x1*7,x2,color='b',linewidth=4,alpha=0.2)
plt.text(15,70000, 'Double in 2 days',alpha=0.5, rotation=55)
plt.text(21,45000, 'Double in 3 days',alpha=0.5, rotation=42)
plt.text(20,7505, 'Double in 4 days',alpha=0.5, rotation=34)
plt.text(19,3000, 'Double in 5 days',alpha=0.5, rotation=29)
plt.text(19,1811, 'Double in 6 days',alpha=0.5, rotation=25)
plt.text(23.5,1724, 'Double in 7 days',alpha=0.5, rotation=20)
plt.plot(x,india,color='r',linewidth=12,marker='o',markersize=15)
plt.savefig('case_100.jpg')
```
